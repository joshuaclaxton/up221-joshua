```python
# distribution across years
year_count = LA_Data['INCIDENT_YEAR'].value_counts()
year_count
# convert into dataframe
year_count = year_count.reset_index()
```

```python
year_count.plot()
```

```python
# give it additional arguments
year_count.plot.line(x = 'year', y = 'count', title = 'Distribution of Reported Shootings by year')
```
